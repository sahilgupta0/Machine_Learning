```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('csvFile/deliveries.csv')
```

# most number of run by players

```python
batman = df.groupby('batsman')
batman['batsman_runs'].sum().sort_values(ascending = False).head(5)
```

# most number of 4

```python
mask = df['batsman_runs']==4
only_four = df[mask]
```

```python
only_four.groupby('batsman')['batsman_runs'].count().sort_values(ascending = False).head()
```

 # most run by virat against team

```python
df.groupby('batsman').get_group('V Kohli').groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending = False).head()
```

# who had hit most run in last 5 over
    condition : played Atleat 200 balls

```python
mask = df['over'] > 15
new_df = df[mask]
```

```python
last5 = new_df.groupby('batsman')
mask2 = lambda  a : a['ball'].count() > 200
moreThan200 = last5.filter(mask2)

moreThan200.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False).head()
# moreThan200
```

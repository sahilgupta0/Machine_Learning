```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
age = [45,-7,23,63,450,62,64,12,56,23,2270,2223,2140,74,45,63,16,181,17,15,-3,21,24,22,27,28,35,93,30]
```

# Before Removing Outliers

```python
sns.histplot(age, kde=True)
```

```python

print(age)
temp = age.copy()
run = 0
age = [x for x in age if x > 0]
while True:
    mean = np.mean(age)
    std = np.std(age)
    print("mean is ", mean, "\nstd is ", std)
    q1 = np.percentile(age, 25)
    q3 = np.percentile(age, 75)
    outliers = []
    for i in age:
        z_score = (i - mean)/std
        if(z_score > 2.5):
            outliers.append(i)
    print("the outliers are " , outliers)
    for item in outliers:
        if item in age:
            age.remove(item)
    if age==temp:
        break
    temp = age.copy()
print("\n\n\nNew filter Age are : ",age)
```

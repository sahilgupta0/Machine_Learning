## Normal example

```python
import numpy as np
import seaborn as sns
from numpy.random import uniform
```

```python
sample_size = 30
num_sample = 1000
shape = (num_sample, sample_size)
sample = uniform(0, 1, shape)
```

```python
sample
```

```python
sample_mean = sample.mean(axis=1)
```

```python
sns.distplot(sample_mean)
```
